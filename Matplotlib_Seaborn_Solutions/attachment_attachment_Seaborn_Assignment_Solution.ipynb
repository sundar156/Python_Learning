```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
cars=pd.read_csv('cars.csv')
```

```python
cars.head()
```

```python
sns.boxplot(cars.Actual_MPG)
```

```python
sns.violinplot(cars.Actual_MPG)
```

### univariate distributions

```python

plt.figure(figsize=(10,7))
sns.distplot(cars.Actual_MPG, bins=20, hist=True, kde=False , rug=False)
```

```python

plt.figure(figsize=(10,7))
sns.distplot(cars.Actual_MPG, bins=20, hist=False, kde=True , rug=False)
```

```python

plt.figure(figsize=(10,7))
sns.distplot(cars.Actual_MPG, bins=20, hist=True, kde=False , rug=True)
```

```python

plt.figure(figsize=(10,7))
sns.distplot(cars.Actual_MPG, bins=20, hist=True, kde=True , rug=True)
```

###  bivariate distributions

```python
sns.jointplot(x="Horsepower", y="Actual_MPG", data=cars)
```

```python
sns.jointplot(x="Horsepower", y="Actual_MPG", data=cars,kind='kde')
```

```python
sns.pairplot(cars)
```

```python
g = sns.PairGrid(cars)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d")
```

```python
d=cars.corr()
plt.figure(figsize=(10,7))
sns.heatmap(d)
```
