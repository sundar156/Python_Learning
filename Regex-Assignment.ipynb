```python
#filepath = 
import os
import codecs
import re
import sys
dirpath = os.getcwd()
print(dirpath)
filepath=dirpath +'/baby1998.html'
#f=codecs.open(filepath)
#re.findall('<tr align='right"><td>1</td><td>")
#print(f.read())
def extract_names(filename):
    names=[]
    f = open(filename, 'r')
    print(f.read())
    text = f.read()
    year_match = re.search('Popularity\\sin\\s(\\d\\d\\d\\d)',text)
    if not year_match:
        try:
            sys.stderr.write('Couldnt find the year!')
          #  sys.exit(1)
        except:
            print('Exception' + sys.exc_info()[0])
            
    year = year_match.group(1)
    names.append(year)
    print(year)
    tuples = re.findall('<td>(\\d+)</td><td>(\\w+)</td>\\<td>(\\w+)</td>', text)
    #print(tuples) 
    names_to_rank={}
    
    for rank_tuple in tuples:
        (rank, boyname, girlname) = rank_tuple  
#        print(rank_tuple)
    if boyname not in names_to_rank:
        names_to_rank[boyname] = rank
    if girlname not in names_to_rank:
        names_to_rank[girlname] = rank
    
    # Get the names, sorted in the right order\n
    sorted_names = sorted(names_to_rank.keys())
   
    # Build up result list, one element per line\n",
    for name in sorted_names:
        names.append(name + "\ "  + names_to_rank[name])
#    print(names_to_rank.values)   
    for name in sorted_names:
        print(name)
# return names
   # print(dict1)
extract_names(filepath)
```
